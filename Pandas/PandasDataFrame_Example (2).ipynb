```python
df_milk['mean']=754.708333
df_milk
```

```python
df_milk['Mean Difference sqaure']= (df_milk["Milk Production"]-df_milk["mean"])**2
df_milk
```

```python
variance=sum(df_milk["Mean Difference sqaure"])/len(df_milk['mean']-1)
variance
```

```python
import numpy as np
std=np.sqrt(variance)
std
```
