```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,110,10)
y=np.arange(20,220,20)
plt.plot(x,y)
plt.xlabel('Number of items')
plt.ylabel('Price of items')
plt.title('item-price graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,110,10)
y=np.arange(20,220,20)
plt.plot(x,y,marker='s',color='#C70039',mec='blue',markersize=10,)
plt.xlabel('Number of items')
plt.ylabel('Price of items')
plt.title('item-price graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,110,10)
y=np.arange(20,220,20)
plt.plot(x,y,marker='s',color='#C70039',mec='blue',markersize=10,linewidth=2,ls=':')
plt.xlabel('Number of items')
plt.ylabel('Price of items')
plt.title('item-price graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,110,10)
y=np.arange(20,220,20)
y1=np.arange(30,330,30)
plt.plot(x,y,marker='s',color='#C70039',mec='blue',markersize=10,linewidth=2,ls=':',label='2020')
plt.plot(x,y1,marker='o',color='#848CE7',mec='m',markersize=10,linewidth=2,ls='-',label='2010')
plt.xlabel('Number of items')
plt.ylabel('Price of items')
plt.title('item-price graph')
plt.legend(loc='best',ncol=2,fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,110,10)
y=np.arange(20,220,20)
plt.plot(x,y,'r*',mec='blue',markersize=10,linewidth=2,ls=':',label='2020')
plt.xlabel('Number of items')
plt.ylabel('Price of items')
plt.title('item-price graph')
plt.legend(loc='best',ncol=2,fontsize=12)
plt.show()

```

```python
HP=[45,60,80,80]
Attack=[49,62,82,100]
Defense=[49,63,83,123]
pokemon_name=['Bulbasaur','Ivysaur','Venusaur','Charmander']
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(pokemon_name,HP,color='red',width=0.45)
ax.bar(pokemon_name,Attack,bottom=HP,width=0.45,color='black')
ax.bar(pokemon_name,Defense,bottom=Attack,width=0.45,color='blue')
ax.set_yticks([30,60,90,120,150])
ax.set_xlabel('pokemon name')
ax.set_ylabel('features of pokemons')
ax.legend(['HP','Attack','Defense'])
plt.show()
```
