```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,110,10)
y=np.arange(20,220,20)
plt.plot(x,y)  # plot a line graph
plt.xlabel('Number of item')
plt.ylabel('Price of item')
plt.title('Item - price graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,110,10)
y=np.arange(20,220,20)
plt.plot(x,y,marker='s',color='#C70039',mec='blue',markersize=10)  # plot a line graph
plt.xlabel('Number of item')
plt.ylabel('Price of item')
plt.title('Item - price graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,110,10)
y=np.arange(20,220,20)
plt.plot(x,y,marker='s',color='#C70039',mec='blue',markersize=10,linewidth=2,ls=':')  # plot a line graph
plt.xlabel('Number of item')
plt.ylabel('Price of item')
plt.title('Item - price graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10,110,10)
y=np.arange(20,220,20)
y1=np.arange(30,330,30)
plt.plot(x,y,marker='s',color='#C70039',mec='blue',markersize=10,linewidth=2,ls=':',label='2020')  # plot a line graph
plt.plot(x,y1,marker='o',color='#848CE7',mec='m',markersize=10,linewidth=2,ls='-',label='2010') 
plt.xlabel('Number of item')
plt.ylabel('Price of item')
plt.title('Item - price graph')
plt.legend(loc='best',ncol=2,fontsize=12)
plt.show()
```

```python
water=[10,20,30,50,60]
oil=[90,80,70,50,40]
gas=[30,80,90,50,60]
companyname=['hp','ongc','rel','servo','indianoil']
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(companyname,water,color='#9999ff',width=0.45)
ax.bar(companyname,oil,bottom=water,width=0.45,color='yellow')
ax.bar(companyname,gas,bottom=oil,width=0.45,color='green')
ax.set_yticks([10,50,100,150])
ax.set_xlabel('Company Name')
ax.set_ylabel('Ratio of water oil gas')
ax.set_title('Companies oil percentage')

ax.legend(['Water','Oil','Gas'])
plt.show()
```
