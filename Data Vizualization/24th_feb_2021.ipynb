```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
bmarks=[10,20,30,50,67,33,44,56,78,85,90,95]
gmarks=[50,58,67,78,98,75,74,20,45,91,46,75]
noofstudents=np.arange(1,13)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(noofstudents,bmarks,marker='o',color='r',s=10)
ax.scatter(noofstudents,gmarks)
ax.legend(['boys marks','girls marks'])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
iris_data=load_iris()
iris_data
```

```python
df_flower=pd.DataFrame(iris_data['data'],columns=iris_data['feature_names'])
df_flower['flowertype']=iris_data['target']
df_flower
```

```python
sns.pairplot(df_flower)
```

```python
plt.scatter(df_flower['petal length (cm)'],df_flower['petal width (cm)'])
```

```python
x=np.arange(1,11)
inr=x * 72
logv=np.log(x)
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax1.plot(x,inr,color='red')
ax1.set_ylabel('INR')
ax1.set_yticks(inr)
ax1.set_xlabel('Dollars')
ax2=ax1.twinx()
ax2.plot(x,logv,ls='--')
ax2.set_yticks(logv)
plt.show()
```
