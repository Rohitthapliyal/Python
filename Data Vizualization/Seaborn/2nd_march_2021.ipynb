```python
import pandas  as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
sns.barplot(x='sex',y='survived',data=df,palette='Greens')
```

```python
sns.countplot(x='class',data=df,palette='Greens')
```

```python
sns.pointplot(x='sex',y='survived')
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
df.head()
```

```python
sns.regplot(x='total_bill',y='tip',data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
df2=pd.get_dummies(df)
df2
```

```python
df2.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df2.corr(),annot=True)
```
